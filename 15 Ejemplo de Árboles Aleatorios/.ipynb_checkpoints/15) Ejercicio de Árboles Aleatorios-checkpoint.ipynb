```python
import pandas            as pd    # Biblioteca para trabajar con DataFrames
import numpy             as np    # Biblioteca para trabajar con àlgebra lineal
import matplotlib.pyplot as plt   # Biblioteca para realizar gràficas
import seaborn           as sns

from sklearn.metrics       import accuracy_score, confusion_matrix, r2_score, explained_variance_score
from sklearn.ensemble      import RandomForestClassifier, RandomForestRegressor
from sklearn.datasets      import make_regression, make_classification
# https://scikit-learn.org/stable/modules/classes.html#module-sklearn.ensemble
# https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets

sns.set()
%matplotlib inline
```

```python
##################################################
# EJEMPLO DE CLASIFICACION
##################################################

# 1) Creamos nuestra atriz de diseño y nuestro vector de clases correctas
matriz_diseño, y_real = make_classification(n_samples=1000,  n_features=4,
                                            n_informative=2, n_redundant=0,
                                            n_classes=2, random_state=0, shuffle=False)

sns.pairplot( data = pd.DataFrame( np.append( arr = matriz_diseño, values = y_real.reshape(1000,1), axis = 1 ),
                                   columns = ["x1","x2","x3","x4","y"] ),
             hue = "y", vars = ["x1","x2","x3","x4"], height = 4, aspect = 1  )
plt.plot()

print("Matriz de Diseño")
print( matriz_diseño[0:5,:] )
print("Y Real")
print( y_real[0:5] )

# 2) Creamos nuestro modelo 
modeloClasificacion = RandomForestClassifier(n_estimators=100, max_depth=2,random_state=0)
print("Parámetros del modelo")
print( modeloClasificacion )

# 3) Ajustamos nuestro modelo
modeloClasificacion.fit(matriz_diseño, y_real)  

# 4) Pronosticamos nuestras clases
y_estimadas = modeloClasificacion.predict( matriz_diseño )

# 5) Revisamos métricas del modelo (importancia de las variables / precision del modelo)
print("Importancia de las variables")
print( modeloClasificacion.feature_importances_ ) # [0.14205973 0.76664038 0.0282433  0.06305659]
print("")
print( accuracy_score(y_real, y_estimadas) )
```

```python
##################################################
# EJEMPLO DE REGRESION
##################################################

# 1) Creamos nuestra atriz de diseño y nuestro vector de clases correctas
matriz_diseño, y_real = make_regression(n_samples=100, n_features=4, noise = 12, bias = 200,
                                        random_state=0, shuffle=False)

print("Matriz de Diseño")
print( matriz_diseño[0:5,:] )
print("Y Real")
print( y_real[0:5] )

sns.pairplot( data = pd.DataFrame( np.append( arr = matriz_diseño, values = y_real.reshape(100,1), axis = 1 ),
                                   columns = ["x1","x2","x3","x4","y"] ), vars = ["x1","x2","x3","x4"], height = 4, aspect = 1  )
plt.show()


# 2) Creamos nuestro modelo 
modeloRegresion = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)

print("Parámetros del modelo")
print( modeloRegresion )

# 3) Ajustamos nuestro modelo
modeloRegresion.fit(matriz_diseño, y_real)  

# 4) Pronosticamos nuestras clases
y_estimadas = modeloRegresion.predict( matriz_diseño )

# 5) Revisamos métricas del modelo (importancia de las variables / precision del modelo)
print("Importancia de las variables")
print( modeloRegresion.feature_importances_ ) # [0.14205973 0.76664038 0.0282433  0.06305659]
print("R cuadrada")
print( r2_score(y_real, y_estimadas) )
print( explained_variance_score(y_real, y_estimadas) )
```

```python
##################################################
# EJEMPLO DE CLASIFICACION
##################################################
datos_iris = pd.read_csv("iris.csv")

# 1) Creamos nuestra atriz de diseño y nuestro vector de clases correctas
matriz_diseño = datos_iris.iloc[:,0:4].values
y_real        = datos_iris.iloc[:,4].values

sns.pairplot( data = datos_iris, hue = "Species", vars = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'],
             height = 4, aspect = 1  )
plt.plot()

print("Matriz de Diseño")
print( matriz_diseño[0:5,:] )
print("Y Real")
print( y_real[0:5] )

# 2) Creamos nuestro modelo 
modeloClasificacion = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=0)
print("Parámetros del modelo")
print( modeloClasificacion )

# 3) Ajustamos nuestro modelo
modeloClasificacion.fit(matriz_diseño, y_real)  

# 4) Pronosticamos nuestras clases
y_estimadas = modeloClasificacion.predict( matriz_diseño )

# 5) Revisamos métricas del modelo (importancia de las variables / precision del modelo)
print("Importancia de las variables")
print( modeloClasificacion.feature_importances_ ) 
print("")
print( accuracy_score(y_real, y_estimadas) )
```

```python
##################################################
# EJEMPLO DE REGRESION
##################################################
datos_4 = pd.read_csv("datos_ajuste_4.csv")
X, Y    = datos_4["X"], datos_4["Y"] 

# Realizamos un grafico de dispersión ( Scatter-Plot )
plt.figure( figsize=(16,6) )                           # Indicamos el tamaño de la gráfica
plt.title("Scatter Plot")                              # Añadimos el titulo
plt.scatter( X, Y , label = "Etiqueta de la serie" )   # Añadimos la 'capa' de puntos 
plt.plot( X, 12 + np.pi*X + 0.5*X*np.cos(2/3*X) , label = "Regresión", color="red" )   # Añadimos la 'capa' de puntos 
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.legend()                                           # Añadimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gráfica

print("R cuadrada")
print( r2_score(Y, 12 + np.pi*X + 0.5*X*np.cos(2/3*X)) )
print( explained_variance_score(Y, 12 + np.pi*X + 0.5*X*np.cos(2/3*X)) )
```

```python
##################################################
# EJEMPLO DE REGRESION
##################################################
matriz_diseño, y_real    = np.array(datos_4["X"]).reshape(100,1), datos_4["Y"] 

print("Matriz de Diseño")
print( matriz_diseño[0:5] )
print("Y Real")
print( y_real[0:5] )

# 2) Creamos nuestro modelo 
modeloRegresion = RandomForestRegressor(random_state=0, n_estimators=100)

print("Parámetros del modelo")
print( modeloRegresion )

# 3) Ajustamos nuestro modelo
modeloRegresion.fit(matriz_diseño, y_real)  

# 4) Pronosticamos nuestras clases
y_estimadas = modeloRegresion.predict( matriz_diseño )

# 5) Revisamos métricas del modelo (importancia de las variables / precision del modelo)
print("Importancia de las variables")
print( modeloRegresion.feature_importances_ ) # [0.14205973 0.76664038 0.0282433  0.06305659]
print("R cuadrada")
print( r2_score(y_real, y_estimadas) )
print( explained_variance_score(y_real, y_estimadas) )

# 6) Realizamos un grafico de dispersión ( Scatter-Plot )
plt.figure( figsize=(16,6) )                           # Indicamos el tamaño de la gráfica
plt.title("Scatter Plot")                              # Añadimos el titulo
plt.scatter( matriz_diseño, y_real , label = "Etiqueta de la serie" )   # Añadimos la 'capa' de puntos 
plt.plot( matriz_diseño, y_estimadas , label = "Random Forest",  color="red" )   # Añadimos la 'capa' de puntos 
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.legend()                                           # Añadimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gráfica
```
